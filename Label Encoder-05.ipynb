# Label Encoder

```python
import pandas as pd
df = pd.DataFrame({'Name' : ['Aman','Nikita','Ajay','OM','Priya','Arpit','Arima']})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit_transform(df['Name'])
```

```python
df['en_name'] = le.fit_transform(df['Name'])
```

```python
df
```

```python
New_data = df
New_data.to_csv('Label_Encoder.csv',index = False)
```
